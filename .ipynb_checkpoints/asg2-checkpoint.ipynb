```python
import pandas as pd
import re
from textblob import TextBlob
```

```python
df = pd.read_csv("../wdps27_final/data/version_015_full.csv",error_bad_lines=False)
df.head(5)
```

```python
def k_to_int(string):
    try:
        s = string.lower().replace(',',".")
        if "k" in s:
            value = float(s.split("k")[0])*1000
            return int(value)
        else:
            return int(s)
    except:
        return string
    
df['votes'] = df.votes.apply(k_to_int)
df = df.sort_values("votes",ascending=False)
df
```

```python
def removePunctuation(line):
    line = str(line)
    line = line.replace("\n"," ").replace("\r"," ")
    if line.strip() == '':
        return ''
    line = re.sub("[^\u4e00-\u9fa5^\s\.\!\:\-\@\#\$\(\)\_\,\;\?^a-z^A-Z^0-9]","",line)
    return line

df["text2"] = df.text.apply(removePunctuation)
df['length'] = df.text2.apply(lambda x: len(x))
df["score"] = df.text2.apply(lambda x: TextBlob(x).sentiment.polarity)
df["subjectivity"] = df.text2.apply(lambda x: TextBlob(x).sentiment.subjectivity)
df[['text','text2','length','score','subjectivity']]
```

```python
def score_range(score):
    s = float(score)
    if s <= -0.5:
        return -2
    elif s <= -0.1:
        return -1
    
    if s >= 0.5:
        return 2
    elif s >= 0.1:
        return 1
    else:
        return 0

df['label'] = df.score.apply(score_range)
```

```python
## remoke subjectivity == 0
df = df[~(df.subjectivity==0)]
```

```python
v = df.label.value_counts().sort_index(ascending=False)
v
```

```python
import matplotlib.pyplot as plt
labels   = v.index.tolist()
quants   = list(v)
```

```python
plt.figure(1, figsize=(7,7))
# For China, make the piece explode a bit
expl = [0.02,0.02,0.1,0.02,0.02]   #第二块即China离开圆心0.1
# Colors used. Recycle if not enough.
colors  = ["#1BFF00","#7DFF44","#FFEC00","#FF5C30","#FF2E00"]  #设置颜色（循环显示）
# Pie Plot
# autopct: format of "percent" string;百分数格式
plt.pie(quants, explode=expl, colors=colors, labels=labels, autopct='%1.1f%%', textprops={'fontsize': 14}, pctdistance=0.7, shadow=True)
plt.title('Sentiment Distribution', bbox={'facecolor':'0.9', 'pad':5})
plt.show()
```
